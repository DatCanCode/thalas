```python
import pandas as pd
```

```python
df = pd.read_csv('data/splited_train_test.csv')
df
```

```python
basic_features = ['SLHC', 'HST', 'HCT', 'MCV', 'MCH', 'MCHC', 'RDWCV', 'SLTC', 'SLBC']
df.dropna(subset=basic_features, inplace=True)
df
```

```python
data = df.copy(deep=True)
for col in basic_features:
    data.loc[:,col] = pd.to_numeric(data[col], errors='coerce', downcast='float')
data
```

```python
data.dropna(subset=basic_features, inplace=True)
data
```

```python
data.info()
```

```python
data_index = data.sort_values("sheet").reset_index(drop=True).reset_index()
data_index
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Train/test proportional from each batch

```python
fig, axs = plt.subplots(ncols=2, figsize=(15, 5))
sns.countplot(data=data_index, x="sheet", hue="train", ax=axs[0])
sns.countplot(data=data_index, x="sheet", hue="train", ax=axs[1])
axs[1].set_yscale('log')
axs[1].set_title("log scale")
```

```python
f = 'SLHC'
IQR = 1.5*(data_index[f].quantile(.75) - data_index[f].quantile(.25))

f_max = data_index[f].quantile(.70) + IQR
f_min = data_index[f].quantile(.25) - IQR

fig, axs = plt.subplots(ncols=2, figsize=(15, 5))
ax = sns.boxplot(data=data_index, y=f, ax=axs[0])
ax.axhline(f_max)
ax.axhline(f_min)
print(f_max, f_min)

m_max = 7.5
m_min = 1.5
ax = sns.scatterplot(data=data_index, x="index", y=f, hue="sheet", edgecolor=[], style='thalas', ax=axs[1])
ax.axhline(f_max), ax.axhline(f_min)
ax.axhline(m_max, color="pink")
ax.axhline(m_min, color="pink")
```

```python
data_index[(data_index[f] >= m_max) | (data_index[f] <= m_min)]
```
