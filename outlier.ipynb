```python
import pandas as pd
```

### Train/test proportional from each batch

```python
fig, axs = plt.subplots(ncols=2, figsize=(15, 5))
sns.countplot(data=data_index, x="sheet", hue="train", ax=axs[0])
sns.countplot(data=data_index, x="sheet", hue="train", ax=axs[1])
axs[1].set_yscale('log')
axs[1].set_title("log scale")
```

```python
def deal_with_outlier(data, col, manual_range, fig_size, trim=False):
    Q1 = data[col].quantile(.25)
    Q3 = data[col].quantile(.75)
    IQR = Q3 - Q1
    f_max = Q3 + 1.5*IQR
    f_min = Q1 - 1.5*IQR
    
    # draw boxplot to get overview of feature (Q1, Q3, outlier)
    fig, axs = plt.subplots(ncols=2, figsize=fig_size)
    ax = sns.boxplot(data=data, y=col, ax=axs[0])
    ax.axhline(f_max)
    ax.axhline(f_min)
    print(f_max, f_min)
    
    
    
    # draw scatter plot with manual range
    m_min, m_max = manual_range
    
    ax = sns.scatterplot(data=data, x="index", y=col, hue="sheet", edgecolor=[], style='thalas', ax=axs[1])
    ax.axhline(f_max), ax.axhline(f_min)
    ax.axhline(m_max, color="pink")
    ax.axhline(m_min, color="pink")
    plt.show(fig)
    
    # filter
    print("Outlier:")
    outlier_mask = (data[col] >= m_max) | (data[col] <= m_min)
    display(data[outlier_mask])
    
    return data[~outlier_mask] if trim else data
```

```python
processed_outlier = deal_with_outlier(data_index, "SLHC", (1.5, 7.5), (20,10), trim=True)
```
